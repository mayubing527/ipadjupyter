## matplotlib的概述

### matplotlib是什么？

1. 在函数设计上参考了Matlab软件，所以以''mat''开头，
2. 中间的''plot'' 表示绘图这一作用
3. 结尾的''lib''表示一个集合，library
4. 所以叫做Matplotlib

### 为什么要使用matplotlib?

1. 绘图功能很强大
2. 你手上有很多数据，但是你不知道如何呈现这些数据
3. 同Python一样，开源，免费

### 绘图的基本流程

 1. 第一部分：通常用于创建画布，以及设计好是否需要创建子图
 2. 第二部分：针对每个子图，或当前画布图像，绘制画布内容
 3. 关于图片的保存和显示

![jupyter](./image/matplotlib.png)

1. **创建画布与创建子图**

- 第一部分主要作用是构建出一张空白的画布，并可以选择是否将整个画布划分为多个部分，方便在同一幅图上绘制多个图形的情况。
- 最简单的绘图可以省略第一部分，而后直接在默认的画布上进行图形绘制。

**num** 接收int、str或Figure，可选，默认是None。 如果具有该标识符的图形已经存在，则将该图形激活并返回。整数表示Figure.number 属性，字符串表示图形标签。如果没有带有标识符的数字或未提供num，则将创建一个新数字，将其激活并返回。如果num是一个int，它将用作Figure.number属性，否则，将使用自动生成的整数值（从1开始，对于每个新数字递增）。如果num是字符串，则将图形标签和窗口标题设置为此值。


**figsize** 接收浮点元组。该参数指定绘图对象宽度，高度，以英寸为单位。默认值：[6.4, 4.8]

**dpi** 指定绘图对象的分辨率，即每英寸多少个像素。如果没提供，默认值100.0


**facecolor颜色**，默认：rcParams["figure.facecolor"]（默认：'white'）
背景颜色。

**edgecolor颜色**，默认：rcParams["figure.edgecolor"]（默认：'white'）
边框颜色。

**frameon** 接收bool值，默认值：True。如果为False，则禁止绘制图形框。

**FigureClass** matplotlib.figure.Figure ，可选，使用自定义Figure实例。

**clear** 接收bool值，默认值：False。如果为True并且该数字已经存在，则将其清除。

**tight_layout** 布尔或字典，默认值：rcParams["figure.autolayout"]（默认值：False）
如果False使用subplotpars。如果True使用tight_layout默认填充调整子图参数。当提供包含密钥的字典pad，w_pad， h_pad，和rect，默认的tight_layout填补处理将被替代。

**constrained_layout** 布尔，默认：rcParams["figure.constrained_layout.use"]（默认：False）
如果True使用约束布局来调整绘图元素的位置。类似于tight_layout，但设计得更加灵活。有关 示例，请参见 约束布局指南。（注意：不适用于add_subplot或 subplot2grid。）

** kwargs**：可选,有关Figure其他可能的参数，请参见https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure 。

```python
import numpy as np
import matplotlib.pyplot as plt
#在[0,10]之间等距取1000个数作为x的取值
x = np.arange(0,2*np.pi,0.01)
y = np.sin(x)
z = np.cos(x)

#1.生成画布
plt.figure(figsize=(6,4))

#2.图表内容绘制
plt.plot(x,y,label="$sin(x)$",color="red",linewidth=3,linestyle='-')
plt.plot(x,z,"b--",label="$cos(x)$")

plt.xlabel("x")#设置x轴标题
plt.ylabel("y")#设置y轴标题
plt.title("PyPlot First Example")# 设置图表标题
#plt.title("绘制的第一张图")
plt.grid()#画格子
plt.ylim(-1.2,1.2)#设置y轴的上、下限

plt.legend()
#3.图表的显示与保存
plt.savefig('sincos.png')
plt.show()
```

### 创建子图

#### 方式一:通过figure的add_subplot

```python
rad = np.arange(0,np.pi*2,0.01)
##第一幅子图
p1 = plt.figure(figsize=(6,4),dpi=80)## 确定画布大小
ax1 = p1.add_subplot(2,2,1)## 创建一个2行2列的子图，并开始在第一行第一列绘制第一幅
plt.title('$x^2$曲线') ## 添加标题
plt.plot(rad,rad**2)## 添加y=x^2曲线
ax2 = p1.add_subplot(2,2,2)## 创建一个2行2列的子图，并开始在第一行第二列绘制第一幅
plt.title('$x^4$曲线') ## 添加标题
plt.plot(rad,rad**4)## 添加y=x^4曲线

##第二行子图
ax3 = p1.add_subplot(2,2,3)## 创建一个2行2列的子图，并开始在第二行第一列绘制第一幅
plt.title('sin 曲线') ## 添加标题
plt.plot(rad,np.sin(rad))## 添加sin曲线
ax4 = p1.add_subplot(2,2,4)## 创建一个2行2列的子图，并开始在第二行第二列绘制第一幅
plt.title('cos 曲线') ## 添加标题
plt.plot(rad,np.cos(rad))## 添加cos曲线
plt.savefig('sc.png')
plt.show()
```

#### 方式二: 通过plt的subplot

```python
rad = np.arange(0,np.pi*2,0.01)
##第一行子图
p1 = plt.figure(figsize=(6,4),dpi=80)## 确定画布大小
ax1 = plt.subplot(2,2,1)## 创建一个2行2列的子图，并开始在第一行第一列绘制第一幅
plt.title('$x^2$曲线') ## 添加标题
plt.plot(rad,rad**2)## 添加y=x^2曲线
ax2 = plt.subplot(2,2,2)## 创建一个2行2列的子图，并开始在第一行第二列绘制第一幅
plt.title('$x^4$曲线') ## 添加标题
plt.plot(rad,rad**4)## 添加y=x^4曲线

##第二行子图
ax3 = plt.subplot(2,2,3)## 创建一个2行2列的子图，并开始在第二行第一列绘制第一幅
plt.title('sin 曲线') ## 添加标题
plt.plot(rad,np.sin(rad))## 添加sin曲线

ax4 = plt.subplot(2,2,4)## 创建一个2行2列的子图，并开始在第二行第二列绘制第一幅
plt.title('cos 曲线') # 添加标题
plt.plot(rad,np.cos(rad))## 添加cos曲线
plt.savefig('sc1.png')
plt.show()
```

#### 方式三: 通过plt的subplots

|<center>subplots 函数|
|-----|
|subplots(nrows = 1，ncols = 1，sharex = False，sharey = False，squeeze = True，subplot_kw = None，gridspec_kw = None，** fig_kw )|
|这个函数返回的是一个元组（tuple）。这个元组中包含了一个figure对象和axes对象集合。|
|**用法** fig, ax = plt.subplots() 把返回的 tuple 进行解包（unpack）成 fig (Figure对象)和 ax (子图对象)两个变量。等价于：|
|fig = plt.figure()；ax = fig.add_subplot(1,1,1)|
|nrows和ncols表示将画布分割成几行几列 例：nrows = 2,ncols = 2表示分割2行2列，起始值都为0，当调用画布中的坐标轴时，ax[0,0]表示调用左上角的，ax[1,1]表示调用右下角的；|
|**sharex**和**sharey**表示坐标轴的属性是否相同，可选的参数：True，False，row，col，默认值均为False，表示画布中的四个ax是相互独立的；|
|** fig_kw:所有其他关键字参数都传递给 figure（）调用。如，设置 figsize=(21, 12) ，则设置了图像大小。|

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(2,2)
fig.suptitle("PyPlot First Example")#设置总标题
rad = np.arange(0,np.pi*2,0.01)

plt.title('$x^2$曲线') ## 添加标题
ax[0][0].plot(rad,rad**2)## 添加y=x^2曲线

plt.title('$x^4$曲线') ## 添加标题
ax[0][1].plot(rad,rad**4)## 添加y=x^4曲线

plt.title('sin 曲线') ## 添加标题
ax[1][0].plot(rad,np.sin(rad))## 添加sin曲线

plt.title('cos 曲线') # 添加标题
ax[1][1].plot(rad,np.cos(rad))## 添加cos曲线
ax[1][1].set(xlabel='x',ylabel='y',title='cos曲线',xlim=(0,2),ylim=(0,2))#设置子图标题
ax[1][1].grid()
plt.savefig('sc1.png')
plt.show()
```

### 注释文本

|<center>无指向型的注释文本plt.pyplot.text()函数|

|补充说明|

family设置字体，自带的可选项有{‘serif’, ‘sans-serif’, ‘cursive’, ‘fantasy’, ‘monospace’}，|

|介绍下其中的几个关键参数：|
|**s** :str,为注释文本内容
**xy** :(float,float), 箭头点所在的坐标位置,二维元组形如(x,y)

### 等高线图：
* 等高线图：contour([X, Y,] Z, [levels], ** kwargs)

|参数|参数值|描述|
|--|--|--|
|X，Y|array-like|可选值Z的坐标|
|Z|array-like（N，M）|绘制轮廓的高度值|
|levels|int或类似数组，可选|确定轮廓线/区域的数量和位置|
||如果int Ñ，使用Ñ数据间隔; 即绘制n + 1个等高线|水平高度自动选择。|
||如果是数组，则在指定的级别绘制轮廓线|值必须按递增顺序排列|

* 若X和Y都是2-D，则其形状必须与Z相同；
* 若X和Y都是1-D，则len(X)==M是Z中的列数，len(Y)==N是Z中的行数。

```python
import numpy as np
import matplotlib.pyplot as plt

delta = 0.25  #网格间距，控制等高线的平滑程度
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)  #生成网格
#print(X,Y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

CS = plt.contour(X, Y, Z)  #绘制等高线图
plt.clabel(CS, inline=1, fontsize=10)  #标出等高值
plt.show()
```
